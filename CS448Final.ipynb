```python
#calculates area of effect from sound source(your location) depending on direction
rt_loc = [1,1]
#ld is a variable to track the direction player came from previously which helps in keeping out of an infite loop
ld = ""
#we know what the maze height is, we need to calculate surface area from you to area in front of you depending on direction you are facing. That is the direction the sound will be emenating towards thus where to mesure reverberations
def area_of_effect(direction, location, maze):
    area_ = 0
    #opening keep track of if we find a way out, if so that is one less surface 
    opening = 0
    if direction == "east":
        inc_ = location[1] + 1
        while maze[location[0]][inc_] == ' ':
            if maze[location[0]][inc_] == '#':
                break
            area_ += 1
            inc_ +=1
            if inc_ == len(maze[0]):
                opening = 1
                break
    if direction == "south":
        inc_ = location[0] + 1
        
        while maze[inc_][location[1]] == ' ':
            
            if maze[inc_][location[1]] == '#':
                break
            area_ += 1
    
            inc_ +=1
            if inc_ == len(maze):
                opening = 1
                break
    if direction == "north":
        inc_ = location[0] - 1
        
        while maze[inc_][location[1]] == ' ':
            
            if maze[inc_][location[1]] == '#':
                break
            area_ += 1
  
            inc_ +=1
            if inc_ == len(maze):
                opening = 1
                break
    if direction == "west":
        inc_ = location[1] - 1
        while maze[location[0]][inc_] == ' ':
            if maze[location[0]][inc_] == '#':
                break
            area_ += 1
            inc_ +=1
            if inc_ == len(maze[0]):
                opening = 1
                break
    return (area_, opening)

#to calculate total room absorbtion, we generate a summmation of each surface area in the 3D space * the respective sound absorbtion coefficient for the respective frequency of sound we generate. For simplicity we will use 500Hz.
#absorbtion coefficients for 500hz:
#painted concrete floor = 0.02
#metal deck(ceiling) = 0.03
#gypsum board(walls) = 0.06

def area_absorbtion(surface_area):
    aa = 0
    #floor
    aa += surface_area[0] * 0.02
    #ceiling 
    aa += surface_area[0] * 0.03
    #flanking walls
    aa += 2* 5* surface_area[0] * 0.06
    #opposite wall
    
    aa += 1* 5 * 0.06
    return aa
#computed the reverberation time with the equation 0.049 * (V/A)
def reverberation_time(surface_area,  absorbtion):
    if absorbtion == 0:
        return 0
    #volume
    vol = 5 * 1 * surface_area[0]
    #formula
    rt = 0.049 * (vol / absorbtion)
    return rt
#this decided best direction of travel. Ideally we should head in direction of MOST reverberation time
def reverb_compass(location, maze, last_dir):
    ladir = last_dir
    #north
    n = area_of_effect("north", location, maze)
    na = area_absorbtion(n)

    nart = reverberation_time(n, na)
    
    #south
    s = area_of_effect("south", location, maze)
    sa = area_absorbtion(s)

    sart = reverberation_time(s, sa)
    
    #east
    e = area_of_effect("east", location, maze)
    ea = area_absorbtion(e)

    eart = reverberation_time(e, ea)
    
    #west
    w = area_of_effect("west", location, maze)
    wa = area_absorbtion(w)
   
    wart = reverberation_time(w, wa)
    
    #trying to find the best direction of travel then traverse in that direction while avoid going back the way we came
    drt = [wart,nart,  eart,sart]
    drt.sort()
    if drt[-1] == nart and ladir != "north":
        print("go south or west?", ladir)
        if ladir == "south":
            if drt[2] == sart:
            
                location[0] += s[0]
                
                ladir = "north"
            if drt[2] == eart:
                location[1] += e[0]
                ladir = "west"
            if drt[2] == wart:
                location[1] -= w[0]
                ladir = "east"
        else:
            
            location[0] -= n[0]
            ladir = "south"
        
    elif drt[-1] == sart :
       
       
        if ladir == "north":
            if drt[2] == nart:
                location[0] -= n[0]
                ladir = "south"
            if drt[2] == eart:
                location[1] += e[0]
                ladir = "west"
            if drt[2] == wart:
                location[1] -= w[0]
                ladir = "east"
        else:
            
    
            location[0] += s[0]
            
            ladir = "north"
    elif drt[-1] == wart :
        
        if ladir == "east":
    
            if drt[2] == sart:
                
                location[0] += s[0]
                ladir = "north"
            if drt[2] == eart:
                
                location[1] += e[0]
                ladir = "west"
            if drt[2] == nart:
                
                location[0] -= n[0]
                ladir = "south"
        else:
            
            location[1] -= w[0]
            ladir = "east"
    elif drt[-1] == eart :
        if ladir == "east":
            if drt[2] == sart:
                location[0] += s[0]
                ladir = "north"
            if drt[2] == wart:
                location[1] -= w[0]
                ladir = "east"
            if drt[2] == nart:
                location[0] -= n[0]
                ladir = "south"
         
        else:
  
            location[1] += e[0]
            ladir = "west"
    
            
                
            
            
            
  
    print("change in location:", location, ladir)
    return ladir

    
    
    
#saze text sample
maze = [
    ['#', '#', '#', '#', '#'],
    ['#', ' ', ' ', ' ', '#'],
    ['#', '#', '#', ' ', '#'],
    ['#', ' ', ' ', ' ', '#'],
    ['#', ' ', '#', '#', '#']
]


#iMaizerunner: f we get within the boundaries of the maze then its essentially game over we escaped
while rt_loc[0] != 0 and rt_loc[0] != 4 and rt_loc[1] != 0 and rt_loc[1] != 4:
    l = reverb_compass(rt_loc, maze, ld)
    #we return direction we traveled from 
    ld = l
    print("again")
#if we get here then that means we found a way out of the maize utilizing reverberation of sound!
```
